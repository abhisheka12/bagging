```python
df = df.sample(569)
df_train = df.iloc[:340,:].sample(100)
df_val = df.iloc[340:455,:].sample(50)
df_test = df.iloc[455:,:].sample(50)
```

```python
X_test = df_val.drop('Cancer',axis=1)
y_test = df_val['Cancer']
```

```python
# 1st Subset of Data for Tree
df_bag = df_train.sample(80,replace=True)

X = df_bag.drop('Cancer',axis=1)
y = df_bag['Cancer']

df_bag
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
classes = ['malignant', 'benign']
def plot_confusionmatrix(pred,y_test,dom):
    print(f'{dom} Confusion matrix')
    cf = confusion_matrix(y_test,pred)
    sns.heatmap(cf,annot=True,yticklabels=classes
               ,xticklabels=classes,cmap='Blues', fmt='g')
    plt.tight_layout()
    plt.show()
```

```python
def evaluate(clf,X,y):
    clf.fit(X,y)
    plot_tree(clf)
    plt.show()
    #plot_decision_regions(X.values, y.values, clf=clf)
    y_pred = clf.predict(X_test)
    print(f'Test score {accuracy_score(y_pred,y_test)}')
    plot_confusionmatrix(y_pred,y_test,dom='Test')
```

```python
dt_bag1 = DecisionTreeClassifier()
```

```python
evaluate(dt_bag1,X,y)
```

```python
# 2nd Subset of Data for Tree
df_bag = df_train.sample(80,replace=True)

X = df_bag.drop('Cancer',axis=1)
y = df_bag['Cancer']

df_bag
dt_bag2 = DecisionTreeClassifier()
evaluate(dt_bag2,X,y)
```

```python
# 3rd Subset of Data for Tree
df_bag = df_train.sample(80,replace=True)

X = df_bag.drop('Cancer',axis=1)
y = df_bag['Cancer']

df_bag
dt_bag3 = DecisionTreeClassifier()
evaluate(dt_bag3,X,y)
```

```python
X_testing = df_test.drop('Cancer',axis=1)
y_testing = df_test['Cancer']
```

```python
print("Predictor 1")
y_predicting = dt_bag1.predict(X_testing)
print(f'Test score {accuracy_score(y_predicting,y_testing)}')
plot_confusionmatrix(y_predicting,y_testing,dom='Test')
```

```python
print("Predictor 2")
y_predicting = dt_bag2.predict(X_testing)
print(f'Test score {accuracy_score(y_predicting,y_testing)}')
plot_confusionmatrix(y_predicting,y_testing,dom='Test') 
```

```python
print("Predictor 3")
y_predicting = dt_bag3.predict(X_testing)
print(f'Test score {accuracy_score(y_predicting,y_testing)}')
plot_confusionmatrix(y_predicting,y_testing,dom='Test') 
```
